```python
dir = '/content/drive/MyDrive/datasets/training'
categories = ['00-damage', '01-whole']
```

```python
training_data = []

def create_training_data():
  for car in categories:
      pth = os.path.join(dir,car)
      output = categories.index(car)
      for img in os.listdir(pth):
          img_arr = cv2.imread(os.path.join(pth,img))
          new_img = cv2.resize(img_arr,(200,200))
          training_data.append([new_img,output])
```

```python
create_training_data()
```

```python
print(len(training_data))
```

```python
tmp = training_data[9][0]
type(tmp)
```

```python
i2 = random.randint(0, 1840)
print(training_data[i2][1])
plt.imshow(training_data[i2][0])
plt.show()
```

```python
random.shuffle(training_data)
```

```python
for sam in training_data:
  print(sam[1])
```

```python
x_train=[]
y_train=[]
for img in training_data:
  x_train.append(img[0])
  y_train.append(img[1])
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
x_train = x_train/255
```

```python
x_train[1]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
model = Sequential([
    Conv2D(128, (3,3), activation='relu', input_shape=(200,200,3)),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics =['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=3 , batch_size= 32)
```

```python
model.save('damage_check_model_V1')
```

```python
new_model = tf.keras.models.load_model('damage_check_model_V1')
```

```python
new_model.evaluate(x_train, y_train)
```
